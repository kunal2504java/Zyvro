```python
import asyncio
from playwright.async_api import async_playwright

async def order_from_zepto(items):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)
        context = await browser.new_context(storage_state="zepto_state.json")
        page = await context.new_page()
        await page.goto("https://www.zeptonow.com", wait_until="domcontentloaded")

        # Handle potential modals (like location or permission popups)
        try:
            await page.wait_for_selector("[data-testid='search-bar-icon']", timeout=15000)
        except:
            print("⚠️ Search bar not visible yet. Checking for popups...")
            # Try closing any popups that might be blocking
            popups = ["button:has-text('Not Now')", "button:has-text('Later')", "button:has-text('OK')"]
            for selector in popups:
                try:
                    await page.click(selector, timeout=2000)
                    print(f"Closed popup: {selector}")
                except:
                    pass
            await page.wait_for_selector("[data-testid='search-bar-icon']", timeout=10000)

        print("🔍 Clicking on search bar...")
        await page.click("[data-testid='search-bar-icon']")
        await page.wait_for_load_state("networkidle")

        # Now we are on the search page with a proper input box
        await page.wait_for_selector("input[type='text']", timeout=10000)

        for item in items:
            print(f"🛒 Searching for {item}...")
            await page.fill("input[type='text']", item)
            await page.keyboard.press("Enter")
            await page.wait_for_timeout(4000)

            try:
                add_buttons = await page.locator("button:has-text('Add')").all()
                if add_buttons:
                    await add_buttons[0].click()
                    print(f"✅ Added {item}")
                else:
                    print(f"⚠️ No 'Add' button found for {item}")
            except Exception as e:
                print(f"⚠️ Error adding {item}: {e}")

            # Return to search for next item
            await page.goto("https://www.zeptonow.com/search", wait_until="domcontentloaded")

        print("🎉 Done adding all items.")
        await context.storage_state(path="zepto_state.json")
        await browser.close()

await order_from_zepto(["milk", "bread"])
```

```python
async def get_browser():
    pw = await async_playwright().start()
    browser = await pw.chromium.launch(headless=False)
    return pw, browser
```

```python
class ZeptoAgent:
    def __init__(self):
        self.state = "zepto_state.json"
        
    async def search_products(self, query: str, limit: int = 10):
        """
        Search for products on Zepto and return detailed product information.

        Use this to find products by name, category, or keywords. Returns product details
        including name, price, ratings, availability, and direct product URLs.

        Args:
            query: Search term (e.g., "dairy milk chocolate", "vegetables", "bread")
            limit: Maximum number of products to return (default: 10, max recommended: 50)

        Returns:
            List of products with: name, variant/size, price, original price, savings,
            rating, review count, product URL, and image URL

        Example:
            results = await search_products("chocolate", limit=5)
            # Returns top 5 chocolate products with prices and details
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print(f"🔍 Searching for: {query}")

        # Navigate with networkidle for faster loading
        await page.goto(f"https://www.zeptonow.com/search?query={query}", wait_until="networkidle")

        # Wait for product cards - reduced timeout
        try:
            await page.wait_for_selector('a[href*="/pn/"]', timeout=5000)
        except:
            print("⚠️ No products found")
            await context.close()
            await browser.close()
            await pw.stop()
            return []

        # Extract data immediately without extra wait
        results = await page.evaluate(f"""
            () => {{
                const cards = document.querySelectorAll('a[href*="/pn/"]');
                const limit = {limit};
                const products = [];

                for (let i = 0; i < Math.min(limit, cards.length); i++) {{
                    const card = cards[i];
                    const url = 'https://www.zeptonow.com' + (card.getAttribute('href') || '');

                    // Fast selectors - direct queries
                    const name = card.querySelector('[data-slot-id="ProductName"] span')?.innerText.trim() || '';
                    const packSize = card.querySelector('[data-slot-id="PackSize"] span')?.innerText.trim() || '';

                    // Price extraction
                    const priceContainer = card.querySelector('[data-slot-id="EdlpPrice"]');
                    const currentPrice = priceContainer?.querySelector('.cptQT7')?.innerText.match(/₹\s*([\d,]+)/);
                    const origPrice = priceContainer?.querySelector('.cx3iWL')?.innerText.match(/₹\s*([\d,]+)/);
                    const savingsText = card.querySelector('.cYCsFo span')?.innerText.match(/₹\s*([\d,]+)/);

                    const price = currentPrice ? parseFloat(currentPrice[1].replace(/,/g, '')) : null;
                    const originalPrice = origPrice ? parseFloat(origPrice[1].replace(/,/g, '')) : null;
                    const savings = savingsText ? parseFloat(savingsText[1].replace(/,/g, '')) : null;

                    // Quick checks
                    const deliveryTime = card.querySelector('[data-slot-id="EtaInformation"]')?.innerText.trim() || null;
                    const hasAdd = card.querySelector('button[data-mode="edlp"]') !== null;
                    const imageUrl = card.querySelector('img')?.src || '';

                    products.push({{
                        index: i,
                        name: name,
                        pack_size: packSize,
                        price: price,
                        original_price: originalPrice,
                        savings: savings,
                        discount_percentage: originalPrice && price ? Math.round(((originalPrice - price) / originalPrice) * 100) : null,
                        rating: null,  // Zepto doesn't show ratings in search
                        reviews: null,
                        has_add_button: hasAdd,
                        in_stock: hasAdd,
                        delivery_time: deliveryTime,
                        url: url,
                        image: imageUrl
                    }});
                }}

                return products;
            }}
        """)

        print(f"🛍 Found {len(results)} products")

        # Quick preview of first 3
        if results:
            for i, p in enumerate(results[:3], 1):
                price_info = f"₹{p['price']}" if p['price'] else "Price N/A"
                if p['savings']:
                    price_info += f" (save ₹{p['savings']})"
                print(f"  {i}. {p['name'][:45]} - {price_info}")

        await context.close()
        await browser.close()
        await pw.stop()

        return results
    
    async def add_to_cart(self, product_urls):
        """
        Add one or multiple products to the shopping cart.

        Takes a product URL or list of URLs and adds one unit of each product to the cart.
        Processes multiple products efficiently by reusing the same browser session.

        Args:
            product_urls: Single product URL (string) or list of product URLs
                         Full URLs like "https://www.zeptonow.com/pn/..." or paths like "/pn/..."

        Returns:
            Dict with success status and list of results for each product

        Example:
            # Single product
            await add_to_cart("https://www.zeptonow.com/pn/dairy-milk/pvid/...")

            # Multiple products
            await add_to_cart([
                "https://www.zeptonow.com/pn/dairy-milk/pvid/...",
                "https://www.zeptonow.com/pn/oreo/pvid/...",
                "/pn/bread/pvid/..."
            ])
        """
        # Handle single URL or list of URLs
        if isinstance(product_urls, str):
            product_urls = [product_urls]

        # Ensure all URLs are full URLs
        full_urls = []
        for url in product_urls:
            if not url.startswith('http'):
                url = f"https://www.zeptonow.com{url}"
            full_urls.append(url)

        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        results = []
        print(f"🛒 Adding {len(full_urls)} product(s) to cart...")

        for idx, product_url in enumerate(full_urls, 1):
            try:
                print(f"  [{idx}/{len(full_urls)}] Opening: {product_url}")
                await page.goto(product_url, wait_until="domcontentloaded")

                # Wait for the add button to appear
                await page.wait_for_selector(
                    'div[aria-label="Add to Cart"] button[aria-label="Increase quantity by 1"]', 
                    timeout=5000
                )

                # Click using JavaScript - fastest method
                clicked = await page.evaluate("""
                    () => {
                        const button = document.querySelector('div[aria-label="Add to Cart"] button[aria-label="Increase quantity by 1"]');
                        if (button) {
                            button.click();
                            return true;
                        }
                        return false;
                    }
                """)

                if clicked:
                    # Short wait for cart update
                    await page.wait_for_timeout(800)
                    print(f"  ✅ Product {idx} added to cart!")
                    results.append({
                        "success": True,
                        "url": product_url,
                        "index": idx - 1
                    })
                else:
                    print(f"  ⚠️ Could not find add button for product {idx}")
                    results.append({
                        "success": False,
                        "url": product_url,
                        "index": idx - 1,
                        "error": "Add button not found"
                    })

            except Exception as e:
                print(f"  ❌ Error adding product {idx}: {e}")
                results.append({
                    "success": False,
                    "url": product_url,
                    "index": idx - 1,
                    "error": str(e)
                })

        # Save session (cart state) once at the end
        await context.storage_state(path=self.state)
        await context.close()
        await browser.close()
        await pw.stop()

        successful = sum(1 for r in results if r["success"])
        print(f"🎉 Successfully added {successful}/{len(full_urls)} products to cart!")

        return {
            "success": successful > 0,
            "total_products": len(full_urls),
            "successful": successful,
            "failed": len(full_urls) - successful,
            "results": results
        }

    async def check_coupons(self, auto_apply: bool = False):
        """
        View all available coupons for the current cart and optionally apply the best one.

        Retrieves all coupons applicable to items in cart, showing which are immediately
        usable and which require additional purchase amount. Can automatically apply
        the first available coupon.

        Args:
            auto_apply: If True, automatically applies the first available coupon (default: False)

        Returns:
            List of coupons with: title, code, applicability status, amount needed to unlock,
            and detailed summary of each coupon's conditions

        Example:
            coupons = await check_coupons(auto_apply=True)
            # Shows all coupons and applies first available one
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print("🧺 Opening cart to check coupons...")
        await page.goto("https://www.zeptonow.com/?cart=open", wait_until="domcontentloaded")

        # Wait for coupon blocks to appear
        await page.wait_for_selector("div[aria-label='Open coupon modal']", timeout=5000)

        # Extract all coupon data in one shot
        coupons = await page.evaluate("""
            () => {
                const blocks = document.querySelectorAll("div[aria-label='Open coupon modal']");
                const results = [];

                for (let i = 0; i < blocks.length; i++) {
                    const card = blocks[i];

                    // Warning text (Shop for ₹xxx more)
                    const warningElem = card.querySelector("div.HKHqS");
                    const warningText = warningElem ? warningElem.innerText.trim() : "";

                    // Extract remaining amount
                    let remainingAmount = null;
                    if (warningText.includes('₹')) {
                        const match = warningText.match(/₹\s*([\d,]+)/);
                        if (match) {
                            remainingAmount = parseInt(match[1].replace(/,/g, ''));
                        }
                    }

                    // Main coupon message
                    const messageElem = card.querySelector(".text-cta2");
                    const message = messageElem ? messageElem.innerText.trim() : "";

                    // Coupon code
                    const codeElem = card.querySelector(".text-body4");
                    let code = "";
                    if (codeElem) {
                        const codeText = codeElem.innerText;
                        const codeMatch = codeText.match(/Code:\s*([A-Z0-9]+)/);
                        code = codeMatch ? codeMatch[1] : codeText.trim();
                    }

                    // Apply button check
                    const applyBtn = card.querySelector("button");
                    const canApply = applyBtn && 
                                    applyBtn.innerText.includes('Apply') && 
                                    !applyBtn.disabled;

                    const status = canApply ? "available_to_apply" 
                                 : remainingAmount ? "locked_requires_more_purchase" 
                                 : "not_applicable";

                    results.push({
                        coupon_index: i,
                        coupon_title: message,
                        coupon_code: code,
                        is_applicable_now: canApply,
                        amount_needed_to_unlock: remainingAmount,
                        coupon_status: status,
                        extracted_summary: `Coupon '${code}' — ${message || 'No message'}; ${
                            canApply ? 'can be applied now.' 
                            : remainingAmount ? `requires ₹${remainingAmount} more purchase.` 
                            : 'currently unavailable.'
                        }`
                    });
                }

                return results;
            }
        """)

        print(f"🎟 Found {len(coupons)} coupons.")

        for coupon in coupons:
            status = 'ACTIVE' if coupon['is_applicable_now'] else \
                     'LOCKED' if coupon['amount_needed_to_unlock'] else 'UNKNOWN'
            print(f"🧾 Coupon #{coupon['coupon_index']+1}: {coupon['coupon_title']} | "
                  f"Code: {coupon['coupon_code']} | Status: {status}")

        # Auto-apply if requested
        if auto_apply:
            for coupon in coupons:
                if coupon['is_applicable_now']:
                    try:
                        print(f"🟢 Applying coupon {coupon['coupon_code']}...")
                        await page.locator("div[aria-label='Open coupon modal']").nth(coupon['coupon_index']).locator("button:has-text('Apply')").click()
                        await page.wait_for_timeout(1500)
                        print("🎉 Coupon applied successfully!")
                        break
                    except Exception as e:
                        print(f"⚠️ Failed to apply coupon: {e}")

        if not coupons:
            print("😕 No coupons detected or coupon UI not visible.")

        await context.close()
        await browser.close()
        await pw.stop()

        return coupons

    async def place_order(self):
        """
        Complete checkout and place the order with payment.

        Processes the entire checkout flow: selects Cash on Delivery (COD) if available,
        falls back to UPI payment if COD unavailable, and confirms the order. Waits for
        order confirmation page and extracts order ID and ETA.

        Returns:
            Order status including: payment method used (COD/UPI), order ID, estimated
            delivery time (ETA), final status, and any errors encountered

        Example:
            result = await place_order()
            # Places order, returns: order_id="019a...", eta="14 mins", status="completed"

        Note: For UPI payments, uses test UPI ID. Real payments require actual UPI confirmation.
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print("🛒 Opening cart checkout page...")
        await page.goto("https://www.zeptonow.com/?cart=open", wait_until="domcontentloaded")

        result = {
            "checkout_started": False,
            "click_to_pay_clicked": False,
            "payment_method": None,
            "verify_clicked": False,
            "final_status": "unknown",
            "order_id": None,
            "eta": None,
            "error": None,
        }

        try:
            # Step 1 — Click 'Click to Pay'
            await page.wait_for_selector("button:has-text('Click to Pay')", timeout=5000)
            click_to_pay_btn = page.locator("button:has-text('Click to Pay')").first

            print("🟢 Clicking 'Click to Pay'...")
            await click_to_pay_btn.click()
            result["click_to_pay_clicked"] = True
            result["checkout_started"] = True

            # Wait for payment page
            await page.wait_for_timeout(3000)

            # Step 2 — Find Juspay frame
            frames = page.frames
            juspay_frame = None
            for frame in frames:
                frame_url = frame.url.lower()
                if "juspay" in frame_url or "payment" in frame_url:
                    juspay_frame = frame
                    break

            frame_context = juspay_frame or page
            print("🧾 Using payment frame context")

            # Step 3 — Check if COD option exists and is available
            cod_option = frame_context.locator('[testid="nvb_cod"]')

            if await cod_option.count() > 0:
                print("🟢 Found COD option, clicking it...")
                await cod_option.first.click()
                result["payment_method"] = "Cash On Delivery"

                await page.wait_for_timeout(1500)

                # Look for "Proceed to Pay" button
                proceed_btn = frame_context.locator('[testid="btn_pay"]')

                if await proceed_btn.count() > 0:
                    print("🟢 Clicking 'Proceed to Pay' for COD...")
                    await proceed_btn.first.click()
                    result["verify_clicked"] = True

                    # Wait for order confirmation page
                    print("⏳ Waiting for order confirmation...")
                    try:
                        # Wait for the order tracking page to load
                        # Look for "Order Tracking" heading or "Arriving in" text
                        await page.wait_for_selector('p:has-text("Order Tracking"), p:has-text("Arriving in")', timeout=10000)

                        # Extract order details from the confirmation page
                        order_info = await page.evaluate("""
                            () => {
                                // Get ETA
                                const etaElement = document.querySelector('[id="eta-timer-content"]');
                                let eta = null;
                                if (etaElement) {
                                    const etaText = etaElement.textContent;
                                    const match = etaText.match(/(\d+)\s*mins?/);
                                    if (match) eta = match[1] + ' mins';
                                }

                                // Get order ID from URL
                                const url = window.location.href;
                                const orderIdMatch = url.match(/order\/status\/([a-f0-9-]+)/);
                                const orderId = orderIdMatch ? orderIdMatch[1] : null;

                                // Get amount to pay
                                const amountElement = document.querySelector('span:has-text("To Pay:")');
                                let amount = null;
                                if (amountElement) {
                                    const amountText = amountElement.parentElement.textContent;
                                    const amountMatch = amountText.match(/₹(\d+)/);
                                    if (amountMatch) amount = amountMatch[1];
                                }

                                return { eta, orderId, amount };
                            }
                        """)

                        result["final_status"] = "checkout_completed"
                        result["order_id"] = order_info.get("orderId")
                        result["eta"] = order_info.get("eta")
                        print(f"✅ Order placed successfully!")
                        print(f"📦 Order ID: {result['order_id']}")
                        print(f"⏱️  ETA: {result['eta']}")

                    except Exception as e:
                        print(f"⚠️ Could not verify order confirmation: {e}")
                        result["final_status"] = "payment_clicked_but_confirmation_uncertain"

                    await context.storage_state(path=self.state)
                    await context.close()
                    await browser.close()
                    await pw.stop()
                    return result
                else:
                    print("⚠️ 'Proceed to Pay' button not found")
            else:
                print("⚠️ COD option not found, trying UPI...")

            # Step 4 — UPI Fallback
            upi_option = frame_context.locator('[testid="nvb_upi"]')

            if await upi_option.count() > 0:
                print("🟢 Selecting UPI payment method...")
                await upi_option.first.click()
                result["payment_method"] = "UPI"

                await page.wait_for_timeout(1500)

                # Enter UPI ID
                upi_input = frame_context.locator('input[placeholder*="upi" i], input[placeholder*="VPA" i]')
                if await upi_input.count() > 0:
                    print("📝 Entering test UPI ID...")
                    await upi_input.first.fill("mithragsn9@oksbi")

                # Click Verify and Pay or Proceed to Pay
                verify_btn = frame_context.locator('button:has-text("Verify and Pay"), button:has-text("Proceed to Pay")')
                if await verify_btn.count() > 0:
                    print("🟢 Clicking payment button...")
                    await verify_btn.first.click()
                    result["verify_clicked"] = True
                    result["final_status"] = "upi_payment_triggered"
                    print("✅ UPI flow triggered!")
                else:
                    result["error"] = "Payment button not found"
            else:
                result["error"] = "No valid payment method found"

        except Exception as e:
            result["error"] = str(e)
            print(f"⚠️ Error during checkout: {e}")

        await context.storage_state(path=self.state)
        await context.close()
        await browser.close()
        await pw.stop()

        return result
    
    async def view_cart(self):
        """
        View all items currently in the shopping cart with quantities and prices.

        Retrieves complete cart contents including individual item details, quantities,
        prices, discounts, and calculates total amount and savings.

        Returns:
            Cart summary with: list of all items (name, variant, quantity, price, savings,
            image), total item count, subtotal amount, total savings, and empty cart status

        Example:
            cart = await view_cart()
            # Returns: total_items=3, subtotal=699, total_savings=31, full item list
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print("🛒 Opening cart...")
        await page.goto("https://www.zeptonow.com/?cart=open", wait_until="domcontentloaded")

        # Wait for cart items to load
        try:
            await page.wait_for_selector('.__6RuoF', timeout=5000)
        except:
            print("📭 Cart appears to be empty or not loaded")
            await context.close()
            await browser.close()
            await pw.stop()
            return {
                "cart_items": [],
                "total_items": 0,
                "subtotal": 0,
                "cart_empty": True
            }

        # Extract all cart data in one shot using JavaScript
        cart_data = await page.evaluate("""
            () => {
                const items = document.querySelectorAll('.__6RuoF');
                const cartItems = [];

                for (let i = 0; i < items.length; i++) {
                    const item = items[i];

                    // Product name
                    const nameElem = item.querySelector('.ngKou');
                    const name = nameElem ? nameElem.innerText.trim() : '';

                    // Product variant/size
                    const variantElem = item.querySelector('.__4SHbK');
                    const variant = variantElem ? variantElem.innerText.trim() : '';

                    // Quantity
                    const qtyElem = item.querySelector('[data-testid$="-cart-qty"]');
                    const quantity = qtyElem ? parseInt(qtyElem.innerText.trim()) : 0;

                    // Current price
                    const priceElem = item.querySelector('.cXVBED');
                    let price = null;
                    if (priceElem) {
                        const priceText = priceElem.innerText.trim();
                        const priceMatch = priceText.match(/₹\s*([\d,]+)/);
                        if (priceMatch) {
                            price = parseFloat(priceMatch[1].replace(/,/g, ''));
                        }
                    }

                    // Original price (if discounted)
                    const originalPriceElem = item.querySelector('.cjLMdz');
                    let originalPrice = null;
                    if (originalPriceElem) {
                        const originalText = originalPriceElem.innerText.trim();
                        const originalMatch = originalText.match(/₹\s*([\d,]+)/);
                        if (originalMatch) {
                            originalPrice = parseFloat(originalMatch[1].replace(/,/g, ''));
                        }
                    }

                    // Image URL
                    const imgElem = item.querySelector('.lmfUk');
                    const imageUrl = imgElem ? imgElem.src : '';

                    // Check if it's a removable item (has Remove button vs quantity controls)
                    const hasRemoveBtn = item.querySelector('.cpG2SV') !== null;
                    const hasQuantityControls = item.querySelector('[data-testid$="-minus-btn"]') !== null;

                    // Calculate savings if original price exists
                    const savings = originalPrice && price ? originalPrice - price : 0;

                    cartItems.push({
                        index: i,
                        name: name,
                        variant: variant,
                        quantity: quantity,
                        price: price,
                        original_price: originalPrice,
                        savings: savings,
                        image_url: imageUrl,
                        is_free_item: hasRemoveBtn && price === 0,
                        can_modify_quantity: hasQuantityControls
                    });
                }

                // Calculate totals
                const subtotal = cartItems.reduce((sum, item) => sum + (item.price * item.quantity), 0);
                const totalSavings = cartItems.reduce((sum, item) => sum + (item.savings * item.quantity), 0);
                const totalItems = cartItems.reduce((sum, item) => sum + item.quantity, 0);

                return {
                    cart_items: cartItems,
                    total_items: totalItems,
                    subtotal: subtotal,
                    total_savings: totalSavings,
                    cart_empty: cartItems.length === 0
                };
            }
        """)

        print(f"🛍️ Found {cart_data['total_items']} items in cart")
        print(f"💰 Subtotal: ₹{cart_data['subtotal']}")
        if cart_data['total_savings'] > 0:
            print(f"🎉 Total savings: ₹{cart_data['total_savings']}")

        # Print individual items
        for item in cart_data['cart_items']:
            free_tag = " [FREE]" if item['is_free_item'] else ""
            savings_tag = f" (Save ₹{item['savings']})" if item['savings'] > 0 else ""
            print(f"  • {item['name']} - {item['variant']} x{item['quantity']} = ₹{item['price'] * item['quantity']}{free_tag}{savings_tag}")

        await context.close()
        await browser.close()
        await pw.stop()

        return cart_data
    
    async def update_cart_quantity(self, item_index: int, action: str = "increase"):
        """
        Increase or decrease the quantity of a specific cart item.

        Modifies quantity by 1 unit. If decreasing quantity of an item with qty=1,
        the item will be removed from cart entirely.

        Args:
            item_index: Position of item in cart (0-based index from view_cart results)
            action: Either "increase" to add 1 unit or "decrease" to remove 1 unit

        Returns:
            Update status with: old quantity, new quantity, and whether item was removed

        Example:
            await update_cart_quantity(item_index=0, action="increase")
            # Increases first item's quantity from 2 to 3
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print(f"🛒 Opening cart to {action} quantity...")
        await page.goto("https://www.zeptonow.com/?cart=open", wait_until="domcontentloaded")

        try:
            await page.wait_for_selector('.__6RuoF', timeout=5000)

            # Get the specific cart item
            cart_items = page.locator('.__6RuoF')
            item = cart_items.nth(item_index)

            # Get current quantity before action
            qty_elem = item.locator('[data-testid$="-cart-qty"]')
            old_qty = int(await qty_elem.inner_text())

            # Click the appropriate button
            if action == "increase":
                button = item.locator('[data-testid$="-plus-btn"]')
                print(f"➕ Increasing quantity from {old_qty}...")
            else:  # decrease
                button = item.locator('[data-testid$="-minus-btn"]')
                print(f"➖ Decreasing quantity from {old_qty}...")

            await button.click()
            await page.wait_for_timeout(800)  # Wait for cart to update

            # Check new quantity (item might be removed if qty was 1 and decreased)
            try:
                new_qty = int(await qty_elem.inner_text())
                print(f"✅ Quantity updated: {old_qty} → {new_qty}")
                result = {
                    "success": True,
                    "item_index": item_index,
                    "old_quantity": old_qty,
                    "new_quantity": new_qty,
                    "removed": False
                }
            except:
                # Item was removed (quantity went to 0)
                print(f"🗑️ Item removed from cart (quantity reached 0)")
                result = {
                    "success": True,
                    "item_index": item_index,
                    "old_quantity": old_qty,
                    "new_quantity": 0,
                    "removed": True
                }

        except Exception as e:
            print(f"❌ Error updating quantity: {e}")
            result = {
                "success": False,
                "error": str(e)
            }

        await context.storage_state(path=self.state)
        await context.close()
        await browser.close()
        await pw.stop()

        return result


    async def remove_from_cart(self, item_index: int):
        """
        Completely remove a specific item from the shopping cart.

        Removes all units of the specified item regardless of current quantity.
        Uses item position from view_cart results.

        Args:
            item_index: Position of item in cart (0-based index from view_cart results)

        Returns:
            Removal status with: item name, quantity removed, and success confirmation

        Example:
            await remove_from_cart(item_index=1)
            # Completely removes the second item from cart
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print(f"🛒 Opening cart to remove item #{item_index}...")
        await page.goto("https://www.zeptonow.com/?cart=open", wait_until="domcontentloaded")

        try:
            await page.wait_for_selector('.__6RuoF', timeout=5000)

            # Get the specific cart item
            cart_items = page.locator('.__6RuoF')
            item = cart_items.nth(item_index)

            # Get item name for logging
            name_elem = item.locator('.ngKou')
            item_name = await name_elem.inner_text()

            # Get current quantity
            qty_elem = item.locator('[data-testid$="-cart-qty"]')
            quantity = int(await qty_elem.inner_text())

            print(f"🗑️ Removing '{item_name}' (quantity: {quantity})...")

            # Click minus button until item is removed
            minus_btn = item.locator('[data-testid$="-minus-btn"]')

            for i in range(quantity):
                await minus_btn.click()
                await page.wait_for_timeout(500)  # Brief wait between clicks
                print(f"  ↳ Clicked minus ({i + 1}/{quantity})")

            print(f"✅ Item removed successfully!")
            result = {
                "success": True,
                "item_index": item_index,
                "item_name": item_name,
                "removed_quantity": quantity
            }

        except Exception as e:
            print(f"❌ Error removing item: {e}")
            result = {
                "success": False,
                "error": str(e)
            }

        await context.storage_state(path=self.state)
        await context.close()
        await browser.close()
        await pw.stop()

        return result


    async def clear_cart(self):
        """
        Remove all items from the shopping cart at once.

        Empties the entire cart by removing all products. This is faster than removing
        items individually and useful for starting fresh or canceling an order.

        Returns:
            Success status and total number of items removed

        Example:
            result = await clear_cart()
            # Returns: items_removed=5, success=True
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print("🛒 Opening cart to clear all items...")
        await page.goto("https://www.zeptonow.com/?cart=open", wait_until="domcontentloaded")

        try:
            # Check if cart is already empty
            empty_cart = await page.locator('text=Your cart is empty').count()

            if empty_cart > 0:
                print("📭 Cart is already empty!")
                result = {
                    "success": True,
                    "items_removed": 0,
                    "message": "Cart was already empty"
                }
            else:
                # Cart has items, proceed to clear
                await page.wait_for_selector('.__6RuoF', timeout=5000)

                # Get all cart items count
                cart_items = page.locator('.__6RuoF')
                item_count = await cart_items.count()

                print(f"🗑️ Found {item_count} items to remove...")

                # Remove all items using JavaScript for speed
                removed_count = await page.evaluate("""
                    () => {
                        const items = document.querySelectorAll('.__6RuoF');
                        let count = 0;

                        items.forEach((item, index) => {
                            // Get quantity
                            const qtyElem = item.querySelector('[data-testid$="-cart-qty"]');
                            const quantity = qtyElem ? parseInt(qtyElem.innerText) : 0;

                            // Get minus button
                            const minusBtn = item.querySelector('[data-testid$="-minus-btn"]');

                            if (minusBtn && quantity > 0) {
                                // Click minus button 'quantity' times with small delays
                                let clicks = 0;
                                const clickInterval = setInterval(() => {
                                    if (clicks < quantity) {
                                        minusBtn.click();
                                        clicks++;
                                    } else {
                                        clearInterval(clickInterval);
                                    }
                                }, 300);  // 300ms between clicks
                                count++;
                            }
                        });

                        return count;
                    }
                """)

                # Wait for all items to be removed (quantity * 300ms * items + buffer)
                wait_time = min(item_count * 1500, 10000)  # Max 10 seconds
                await page.wait_for_timeout(wait_time)

                # Verify cart is empty
                empty_check = await page.locator('text=Your cart is empty').count()

                if empty_check > 0:
                    print(f"✅ Successfully cleared {removed_count} items from cart!")
                    result = {
                        "success": True,
                        "items_removed": removed_count,
                        "message": "Cart cleared successfully"
                    }
                else:
                    print(f"⚠️ Cleared items but cart may still have some products")
                    result = {
                        "success": True,
                        "items_removed": removed_count,
                        "message": "Partial clear - some items may remain"
                    }

        except Exception as e:
            print(f"❌ Error clearing cart: {e}")
            result = {
                "success": False,
                "error": str(e),
                "items_removed": 0
            }

        await context.storage_state(path=self.state)
        await context.close()
        await browser.close()
        await pw.stop()

        return result
    
    async def get_orders(self, limit: int = 3):
        """
        Retrieve order history with detailed product information.

        Fetches recent orders including status, items ordered, prices, and timestamps.
        Automatically loads more orders if requested limit exceeds initially visible orders.
        Each order includes full product details with names, quantities, and prices.

        Args:
            limit: Number of recent orders to retrieve (default: 5, recommended max: 20)

        Returns:
            List of orders with: order ID, status (delivered/cancelled/processing), placement
            date/time, total amount, list of products ordered (with names, quantities, prices),
            product images, and available actions (reorder, rate, etc.)

        Example:
            orders = await get_orders(limit=10)
            # Returns last 10 orders with full details including all products in each order
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        print(f"📦 Fetching orders (limit: {limit})...")
        await page.goto("https://www.zeptonow.com/account/orders", wait_until="domcontentloaded")

        try:
            # Wait for orders to load
            await page.wait_for_selector('a[href*="/order/"]', timeout=5000)

            # Click "Load More" button if we need more orders than initially loaded
            current_count = 0
            while current_count < limit:
                orders = page.locator('a[href*="/order/"]')
                current_count = await orders.count()

                if current_count >= limit:
                    break

                load_more_btn = page.locator('button[aria-label="Load More"]')
                if await load_more_btn.count() > 0:
                    print(f"  ↳ Loading more orders... (currently {current_count})")
                    await load_more_btn.click()
                    await page.wait_for_timeout(1500)
                else:
                    print(f"  ↳ No more orders available (total: {current_count})")
                    break

            # Extract order URLs and basic info
            orders_basic = await page.evaluate(f"""
                (limit) => {{
                    // The <a> tag wraps the entire order card
                    const orderLinks = document.querySelectorAll('a[href*="/order/"]');
                    const orders = [];

                    for (let i = 0; i < Math.min(limit, orderLinks.length); i++) {{
                        const link = orderLinks[i];
                        const orderUrl = link.getAttribute('href') || '';

                        // Extract order ID from href
                        let orderId = '';
                        if (orderUrl) {{
                            const parts = orderUrl.split('/').filter(p => p);
                            // Get the last part, remove query params
                            orderId = parts[parts.length - 1].split('?')[0];
                        }}

                        // Get the card content (child of the link)
                        const card = link;

                        // Order status
                        const statusElem = card.querySelector('.text-heading6');
                        const status = statusElem ? statusElem.innerText.trim() : '';

                        // Placed at date/time
                        const dateElem = card.querySelector('.text-body2');
                        const placedAt = dateElem ? dateElem.innerText.replace('Placed at ', '').trim() : '';

                        // Total amount
                        const amountElem = card.querySelector('.text-heading5');
                        let amount = null;
                        if (amountElem) {{
                            const amountText = amountElem.innerText.trim();
                            const amountMatch = amountText.match(/₹([\\d,.]+)/);
                            if (amountMatch) {{
                                amount = parseFloat(amountMatch[1].replace(/,/g, ''));
                            }}
                        }}

                        // Product images
                        const images = [];
                        const imgElements = card.querySelectorAll('.no-scrollbar img');
                        imgElements.forEach(img => {{
                            if (img.src) images.push(img.src);
                        }});

                        // Status type
                        let statusType = 'unknown';
                        const statusLower = status.toLowerCase();
                        if (statusLower.includes('delivered')) statusType = 'delivered';
                        else if (statusLower.includes('cancelled')) statusType = 'cancelled';
                        else if (statusLower.includes('processing') || statusLower.includes('packing')) statusType = 'processing';
                        else if (statusLower.includes('transit') || statusLower.includes('way')) statusType = 'in_transit';

                        // Available actions
                        const buttons = card.querySelectorAll('button .text-cta2');
                        const actions = [];
                        buttons.forEach(btn => {{
                            const text = btn.innerText.trim();
                            if (text) actions.push(text);
                        }});

                        orders.push({{
                            index: i,
                            order_id: orderId,
                            status: status,
                            status_type: statusType,
                            placed_at: placedAt,
                            amount: amount,
                            product_images: images,
                            available_actions: actions,
                            order_url: orderUrl ? 'https://www.zeptonow.com' + orderUrl : '',
                            is_active: statusType === 'processing' || statusType === 'in_transit'
                        }});
                    }}

                    return orders;
                }}
            """, limit)

            print(f"✅ Found {len(orders_basic)} orders")

            # Fetch detailed product info for each order
            print("📋 Fetching order details...")
            for order in orders_basic:
                if order['order_id']:
                    try:
                        # Navigate to order details page
                        detail_url = f"https://www.zeptonow.com/order/{order['order_id']}"
                        await page.goto(detail_url, wait_until="domcontentloaded")
                        await page.wait_for_timeout(1000)

                        # Extract product details
                        products = await page.evaluate("""
                            () => {
                                const productCards = document.querySelectorAll('.DqiMl');
                                const products = [];

                                for (let card of productCards) {
                                    // Product name
                                    const nameElem = card.querySelector('.XTK8s');
                                    const name = nameElem ? nameElem.innerText.trim() : '';

                                    // Product variant/size
                                    const variantElem = card.querySelector('.QJHbA');
                                    const variant = variantElem ? variantElem.innerText.trim() : '';

                                    // Quantity
                                    const qtyElem = card.querySelector('.Oz3g8');
                                    const quantityText = qtyElem ? qtyElem.innerText.trim() : '';
                                    const qtyMatch = quantityText.match(/(\\d+)\\s*unit/);
                                    const quantity = qtyMatch ? parseInt(qtyMatch[1]) : 1;

                                    // Price
                                    const priceElem = card.querySelector('.kF_hN');
                                    let price = null;
                                    if (priceElem) {
                                        const priceText = priceElem.innerText.trim();
                                        const priceMatch = priceText.match(/₹([\\d,.]+)/);
                                        if (priceMatch) {
                                            price = parseFloat(priceMatch[1].replace(/,/g, ''));
                                        }
                                    }

                                    // Original price (if discounted)
                                    const originalPriceElem = card.querySelector('.L5X_W');
                                    let originalPrice = null;
                                    if (originalPriceElem) {
                                        const originalText = originalPriceElem.innerText.trim();
                                        const originalMatch = originalText.match(/₹([\\d,.]+)/);
                                        if (originalMatch) {
                                            originalPrice = parseFloat(originalMatch[1].replace(/,/g, ''));
                                        }
                                    }

                                    // Product image
                                    const imgElem = card.querySelector('img');
                                    const image = imgElem ? imgElem.src : '';

                                    // Product URL
                                    const linkElem = card.querySelector('a');
                                    const productUrl = linkElem ? linkElem.getAttribute('href') : '';

                                    if (name) {
                                        products.push({
                                            name: name,
                                            variant: variant,
                                            quantity: quantity,
                                            price: price,
                                            original_price: originalPrice,
                                            savings: originalPrice && price ? originalPrice - price : 0,
                                            image: image,
                                            product_url: productUrl ? 'https://www.zeptonow.com' + productUrl : ''
                                        });
                                    }
                                }

                                return products;
                            }
                        """)

                        order['products'] = products
                        order['total_items'] = len(products)
                        print(f"  ✓ Order {order['order_id'][:8]}... - {len(products)} products")

                    except Exception as e:
                        print(f"  ⚠️ Could not fetch details for order {order['order_id']}: {e}")
                        order['products'] = []
                        order['total_items'] = 0
                else:
                    order['products'] = []
                    order['total_items'] = 0

            # Print order summary
            for order in orders_basic:
                status_emoji = {
                    'delivered': '✅',
                    'cancelled': '❌',
                    'processing': '📦',
                    'in_transit': '🚚',
                    'unknown': '❓'
                }.get(order['status_type'], '❓')

                items_text = f" - {order.get('total_items', '?')} items"
                print(f"  {status_emoji} {order['status']} | ₹{order['amount']} | {order['placed_at']}{items_text}")

            result = {
                "success": True,
                "orders": orders_basic,
                "total_orders": len(orders_basic)
            }

        except Exception as e:
            print(f"❌ Error fetching orders: {e}")
            result = {
                "success": False,
                "error": str(e),
                "orders": []
            }

        await context.close()
        await browser.close()
        await pw.stop()

        return result
    
    async def get_product_details(self, product_url: str):
        """
        Get comprehensive information about a specific product.

        Retrieves all available product details including specifications, pricing, nutrition
        facts, ingredients, seller information, available coupons, and customer reviews.

        Args:
            product_url: Full Zepto product URL or path (from search results or order history)

        Returns:
            Complete product information including: name, brand, net quantity, rating, review
            count, delivery time, regular and super-saver prices, available coupons, detailed
            highlights (ingredients, allergens, nutrition, storage), and seller information

        Example:
            details = await get_product_details("/pn/dairy-milk/pvid/...")
            # Returns full product specs, nutrition info, 5+ available coupons, etc.

        Use this to: check product details before purchase, verify ingredients/allergens,
        compare prices with super-saver offers, or view all applicable coupons.
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        # Handle both full URLs and paths
        if not product_url.startswith('http'):
            product_url = f"https://www.zeptonow.com{product_url}"

        print(f"📦 Fetching product details...")
        await page.goto(product_url, wait_until="domcontentloaded")

        try:
            # Wait for product details to load
            await page.wait_for_selector('h1', timeout=5000)

            # Extract all product data in one shot
            product_data = await page.evaluate("""
                () => {
                    // Basic Info
                    const nameElem = document.querySelector('h1');
                    const name = nameElem ? nameElem.innerText.trim() : '';

                    const brandElem = document.querySelector('.EwqK4 .ccJTKY');
                    const brand = brandElem ? brandElem.innerText.trim() : '';

                    // Net Quantity
                    const netQtyElem = document.querySelector('.BoqfC .cpyBgD');
                    let netQty = '';
                    if (netQtyElem) {
                        const text = netQtyElem.innerText;
                        const match = text.match(/Net Qty:\\s*(.+)/);
                        netQty = match ? match[1].trim() : '';
                    }

                    // Rating
                    const ratingElem = document.querySelector('.NFZ3f .FnCU0');
                    const rating = ratingElem ? parseFloat(ratingElem.innerText) : null;

                    const reviewCountElem = document.querySelector('.ti0_P span:last-child');
                    let reviewCount = null;
                    if (reviewCountElem) {
                        const text = reviewCountElem.innerText.replace(/[()]/g, '');
                        if (text.includes('k')) {
                            reviewCount = parseFloat(text.replace('k', '')) * 1000;
                        } else {
                            reviewCount = parseInt(text);
                        }
                    }

                    // Delivery Time
                    const etaElem = document.querySelector('#eta-strip .cey2Nb');
                    const deliveryTime = etaElem ? etaElem.innerText.trim() : null;

                    // Price
                    const priceElem = document.querySelector('.TvBIc');
                    let price = null;
                    if (priceElem) {
                        const priceText = priceElem.innerText.trim();
                        const match = priceText.match(/₹([\\d,.]+)/);
                        if (match) {
                            price = parseFloat(match[1].replace(/,/g, ''));
                        }
                    }

                    // Super Saver Price
                    const superSaverElem = document.querySelector('.eG_bL');
                    let superSaverPrice = null;
                    if (superSaverElem) {
                        const text = superSaverElem.innerText.trim();
                        const match = text.match(/₹([\\d,.]+)/);
                        if (match) {
                            superSaverPrice = parseFloat(match[1].replace(/,/g, ''));
                        }
                    }

                    // Coupons
                    const coupons = [];
                    const couponButtons = document.querySelectorAll('button[data-couponid]');
                    couponButtons.forEach(btn => {
                        const text = btn.querySelector('.__0fsT_ p');
                        if (text) {
                            const imgElem = btn.querySelector('img');
                            coupons.push({
                                title: text.innerText.trim(),
                                image: imgElem ? imgElem.src : '',
                                id: btn.getAttribute('data-couponid')
                            });
                        }
                    });

                    // Highlights (Product Details)
                    const highlights = {};
                    const highlightRows = document.querySelectorAll('#productHighlights .KjTQZ');
                    highlightRows.forEach(row => {
                        const cells = row.querySelectorAll('div');
                        if (cells.length >= 2) {
                            const keyElem = cells[0].querySelector('h3');
                            const valueElem = cells[1].querySelector('p');
                            if (keyElem && valueElem) {
                                const key = keyElem.innerText.trim();
                                const value = valueElem.innerText.trim();
                                highlights[key] = value;
                            }
                        }
                    });

                    // Information (Seller details, etc.)
                    const information = {};
                    const infoRows = document.querySelectorAll('#productInformationL4 .KjTQZ');
                    infoRows.forEach(row => {
                        const cells = row.querySelectorAll('div');
                        if (cells.length >= 2) {
                            const keyElem = cells[0].querySelector('h3');
                            const valueElem = cells[1].querySelector('p');
                            if (keyElem && valueElem) {
                                const key = keyElem.innerText.trim();
                                const value = valueElem.innerText.trim();
                                information[key] = value;
                            }
                        }
                    });

                    // Product Image
                    const mainImageElem = document.querySelector('.u-relative.INlj0 img');
                    const image = mainImageElem ? mainImageElem.src : '';

                    return {
                        name: name,
                        brand: brand,
                        net_quantity: netQty,
                        rating: rating,
                        review_count: reviewCount,
                        delivery_time: deliveryTime,
                        price: price,
                        super_saver_price: superSaverPrice,
                        savings: superSaverPrice && price ? price - superSaverPrice : 0,
                        image: image,
                        coupons: coupons,
                        highlights: highlights,
                        information: information,
                        url: window.location.href
                    };
                }
            """)

            print(f"✅ Product: {product_data['name']}")
            print(f"   Brand: {product_data['brand']}")
            print(f"   Price: ₹{product_data['price']}")
            if product_data['super_saver_price']:
                print(f"   Super Saver: ₹{product_data['super_saver_price']} (Save ₹{product_data['savings']})")
            print(f"   Rating: {product_data['rating']}⭐ ({product_data['review_count']} reviews)")
            print(f"   Delivery: {product_data['delivery_time']}")
            print(f"   Available Coupons: {len(product_data['coupons'])}")

            result = {
                "success": True,
                "product": product_data
            }

        except Exception as e:
            print(f"❌ Error fetching product details: {e}")
            result = {
                "success": False,
                "error": str(e)
            }

        await context.close()
        await browser.close()
        await pw.stop()

        return result
```

```python
class AmazonAgent:
    def __init__(self):
        self.state = "amazon_state.json"
        
    async def search_products(self, query: str, limit: int = 10):
        """
        Search for products on Amazon India and return product details.
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        # Navigate to Amazon search
        search_url = f"https://www.amazon.in/s?k={query.replace(' ', '+')}"
        await page.goto(search_url, wait_until="domcontentloaded")

        # Wait for products to load
        await page.wait_for_selector('[data-component-type="s-search-result"]', timeout=10000)

        # Debug: Let's see what selectors actually work
        debug_info = await page.evaluate("""
            () => {
                const card = document.querySelector('[data-component-type="s-search-result"]');
                if (!card) return { error: 'No card found' };

                return {
                    h2_a_span: card.querySelector('h2 a span') ? 'EXISTS' : 'MISSING',
                    h2_span: card.querySelector('h2 span') ? 'EXISTS' : 'MISSING',
                    h2_text: card.querySelector('h2') ? card.querySelector('h2').innerText.substring(0, 100) : 'NO H2',
                    price_whole: card.querySelector('.a-price-whole') ? 'EXISTS' : 'MISSING',
                    rating_small: card.querySelector('.a-icon-star-small') ? 'EXISTS' : 'MISSING',
                    all_classes: Array.from(card.querySelectorAll('[class*="price"]')).map(el => el.className).slice(0, 5)
                };
            }
        """)
        print("🔍 Debug info:", debug_info)

        # Get all product data
        results = await page.evaluate(f"""
            () => {{
                const cards = document.querySelectorAll('[data-component-type="s-search-result"]');
                const limit = {limit};
                const products = [];

                for (let i = 0; i < Math.min(limit, cards.length); i++) {{
                    const card = cards[i];

                    // Name - try multiple selectors
                    let name = '';
                    const nameSelectors = [
                        'h2 a span',
                        'h2 span',
                        'h2 a',
                        '.a-size-medium.a-color-base.a-text-normal',
                        '[data-cy="title-recipe"] span'
                    ];
                    for (const selector of nameSelectors) {{
                        const elem = card.querySelector(selector);
                        if (elem && elem.innerText.trim()) {{
                            name = elem.innerText.trim();
                            break;
                        }}
                    }}

                    // URL
                    const linkElem = card.querySelector('h2 a') || card.querySelector('a.a-link-normal');
                    let url = linkElem ? linkElem.getAttribute('href') : '';
                    if (url && !url.startsWith('http')) {{
                        url = 'https://www.amazon.in' + url;
                    }}

                    // Price - try multiple selectors
                    let price = null;
                    let priceText = '';
                    const priceSelectors = [
                        '.a-price-whole',
                        '.a-price .a-offscreen',
                        '[data-a-color="price"] .a-offscreen'
                    ];
                    for (const selector of priceSelectors) {{
                        const elem = card.querySelector(selector);
                        if (elem) {{
                            priceText = elem.innerText.trim();
                            const priceClean = priceText.replace(/[,₹]/g, '').trim();
                            if (priceClean) {{
                                price = parseFloat(priceClean);
                                break;
                            }}
                        }}
                    }}

                    // Rating
                    let rating = null;
                    const ratingElem = card.querySelector('.a-icon-star-small .a-icon-alt') || 
                                       card.querySelector('i[class*="a-star"] span');
                    if (ratingElem) {{
                        const ratingText = ratingElem.innerText || ratingElem.textContent;
                        const ratingMatch = ratingText.match(/(\d+\.?\d*)/);
                        rating = ratingMatch ? parseFloat(ratingMatch[1]) : null;
                    }}

                    // Review count
                    let reviews = null;
                    const reviewElem = card.querySelector('span[aria-label*="ratings"]') ||
                                       card.querySelector('span[aria-label*="stars"]');
                    if (reviewElem) {{
                        const reviewLabel = reviewElem.getAttribute('aria-label') || '';
                        const reviewMatch = reviewLabel.match(/([\d,]+)/);
                        reviews = reviewMatch ? parseInt(reviewMatch[1].replace(/,/g, '')) : null;
                    }}

                    // ASIN
                    const asin = card.getAttribute('data-asin') || '';

                    // Prime
                    const isPrime = card.querySelector('i[aria-label="Amazon Prime"]') !== null ||
                                    card.querySelector('[aria-label="Amazon Prime"]') !== null;

                    products.push({{
                        index: i,
                        asin: asin,
                        name: name,
                        price: price,
                        price_text: priceText,
                        rating: rating,
                        reviews: reviews,
                        is_prime: isPrime,
                        url: url
                    }});
                }}

                return products;
            }}
        """)

        print(f"🛍 Found {len(results)} products")

        await context.close()
        await browser.close()
        await pw.stop()

        return results
    
    async def view_product(self, product_url: str):
        """
        View essential product information from an Amazon product page.

        Retrieves key product details including title, price, rating, availability,
        delivery info, and customization options (color, size, storage, etc.).

        Args:
            product_url: Full Amazon product URL or ASIN
                         (e.g., "https://www.amazon.in/dp/B0CHX1W1XY" or "B0CHX1W1XY")

        Returns:
            Dict with essential product information including pricing, availability,
            variants, delivery details, and seller information

        Example:
            product = await view_product("https://www.amazon.in/dp/B0CHX1W1XY")
            # Returns: name, price, rating, colors, sizes, delivery time, stock status
        """
        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        # Handle ASIN or full URL (including tracking URLs)
        if not product_url.startswith('http'):
            # Just an ASIN
            product_url = f"https://www.amazon.in/dp/{product_url}"
        else:
            # Extract ASIN from URL (handles tracking URLs too)
            import re
            asin_match = re.search(r'/dp/([A-Z0-9]{10})', product_url)
            if asin_match:
                asin = asin_match.group(1)
                product_url = f"https://www.amazon.in/dp/{asin}"

        print(f"📦 Viewing product...")
        await page.goto(product_url, wait_until="domcontentloaded")

        try:
            # Wait for product title to load
            await page.wait_for_selector('#productTitle', timeout=5000)

            # Extract all essential product data in one shot
            product_data = await page.evaluate("""
                () => {
                    // Product Title
                    const titleElem = document.querySelector('#productTitle');
                    const title = titleElem ? titleElem.innerText.trim() : '';

                    // Brand
                    const brandElem = document.querySelector('#bylineInfo');
                    const brand = brandElem ? brandElem.innerText.replace('Visit the ', '').replace(' Store', '').trim() : '';

                    // Rating
                    const ratingElem = document.querySelector('.a-icon-star span');
                    let rating = null;
                    if (ratingElem) {
                        const ratingText = ratingElem.innerText;
                        const match = ratingText.match(/([\\d.]+)/);
                        rating = match ? parseFloat(match[1]) : null;
                    }

                    // Review count
                    const reviewElem = document.querySelector('#acrCustomerReviewText');
                    let reviewCount = null;
                    if (reviewElem) {
                        const text = reviewElem.innerText.replace(/,/g, '');
                        const match = text.match(/([\\d]+)/);
                        reviewCount = match ? parseInt(match[1]) : null;
                    }

                    // Price from corePriceDisplay_desktop - multiple attempts
                    let price = null;
                    let originalPrice = null;
                    let discount = null;

                    // Try corePriceDisplay_desktop first
                    const priceBlock = document.querySelector('#corePriceDisplay_desktop_feature_div');
                    if (priceBlock) {
                        const priceWhole = priceBlock.querySelector('.a-price-whole');
                        if (priceWhole) {
                            price = parseFloat(priceWhole.innerText.replace(/,/g, ''));
                        }

                        // MRP from basisPrice
                        const mrpElem = priceBlock.querySelector('.basisPrice .a-offscreen');
                        if (mrpElem) {
                            const mrpText = mrpElem.innerText;
                            const match = mrpText.match(/₹([\\d,]+)/);
                            if (match) {
                                originalPrice = parseFloat(match[1].replace(/,/g, ''));
                            }
                        }

                        // Discount from savingsPercentage
                        const discountElem = priceBlock.querySelector('.savingsPercentage');
                        if (discountElem) {
                            const match = discountElem.innerText.match(/([\\d]+)%/);
                            discount = match ? parseInt(match[1]) : null;
                        }
                    }

                    // Exchange offer from buyBackAccordionRow
                    let exchangeOffer = null;
                    const exchangeElem = document.querySelector('#maxBuyBackDiscountSection .a-color-price');
                    if (exchangeElem) {
                        exchangeOffer = exchangeElem.innerText.trim();
                    }

                    // Stock status
                    const stockElem = document.querySelector('#availability span');
                    const stockStatus = stockElem ? stockElem.innerText.trim() : 'Unknown';
                    const inStock = stockStatus.toLowerCase().includes('in stock');

                    // Delivery info from deliveryBlock
                    let deliveryTime = null;
                    const deliveryElem = document.querySelector('#mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE');
                    if (deliveryElem) {
                        deliveryTime = deliveryElem.innerText.trim();
                    }

                    // Deliver to location from contextualIngressPt
                    let deliverTo = null;
                    const locationElem = document.querySelector('#contextualIngressPtLabel');
                    if (locationElem) {
                        deliverTo = locationElem.innerText.trim();
                    }

                    // Ships from / Sold by - proper selectors
                    let shipsFrom = null;
                    let soldBy = null;

                    // Ships from
                    const shipsFromElem = document.querySelector('[offer-display-feature-name="desktop-fulfiller-info"] .offer-display-feature-text-message');
                    if (shipsFromElem) {
                        shipsFrom = shipsFromElem.innerText.trim();
                    }

                    // Sold by
                    const soldByElem = document.querySelector('[offer-display-feature-name="desktop-merchant-info"] .offer-display-feature-text-message a, [offer-display-feature-name="desktop-merchant-info"] .offer-display-feature-text-message');
                    if (soldByElem) {
                        soldBy = soldByElem.innerText.trim();
                    }

                    // Payment info
                    let paymentInfo = null;
                    const paymentElem = document.querySelector('[offer-display-feature-name="desktop-dynamic-secure-transaction"] .offer-display-feature-text-message');
                    if (paymentElem) {
                        paymentInfo = paymentElem.innerText.trim();
                    }

                    // Gift options
                    let giftOptions = null;
                    const giftElem = document.querySelector('[offer-display-feature-name="desktop-dynamic-gift-wrap"] .offer-display-feature-text-message, [offer-display-feature-name="desktop-dynamic-gift-wrap"] .gift-wrap-message');
                    if (giftElem) {
                        giftOptions = giftElem.innerText.trim();
                    }

                    // Color variants with prices from inline-twister-swatch
                    const colors = [];
                    const colorElems = document.querySelectorAll('li.inline-twister-swatch[data-asin]');
                    colorElems.forEach(li => {
                        // Check if this is a color variant (has image)
                        const img = li.querySelector('.swatch-image');
                        if (!img) return;

                        const isSelected = li.getAttribute('data-initiallyselected') === 'true';
                        const isUnavailable = li.getAttribute('data-initiallyunavailable') === 'true';
                        const asin = li.getAttribute('data-asin');

                        colors.push({
                            name: img.alt || '',
                            selected: isSelected,
                            available: !isUnavailable,
                            asin: asin,
                            image: img.src
                        });
                    });

                    // Size/Storage variants from text swatches
                    const sizes = [];
                    const sizeElems = document.querySelectorAll('li.swatch-list-item-text[data-asin]');
                    sizeElems.forEach(li => {
                        const isSelected = li.getAttribute('data-initiallyselected') === 'true';
                        const isUnavailable = li.getAttribute('data-initiallyunavailable') === 'true';
                        const asin = li.getAttribute('data-asin');

                        // Get size name from text
                        const sizeTextElem = li.querySelector('.swatch-title-text-display');
                        const sizeName = sizeTextElem ? sizeTextElem.innerText.trim() : '';

                        if (sizeName) {
                            sizes.push({
                                name: sizeName,
                                selected: isSelected,
                                available: !isUnavailable,
                                asin: asin
                            });
                        }
                    });

                    // Fallback: If no sizes found from swatches, try dropdown
                    if (sizes.length === 0) {
                        const dropdownOptions = document.querySelectorAll('#native_dropdown_selected_size_name option');
                        dropdownOptions.forEach(option => {
                            const sizeName = option.innerText.trim();
                            if (sizeName && sizeName !== 'Select') {
                                sizes.push({
                                    name: sizeName,
                                    selected: option.selected,
                                    available: !option.disabled,
                                    asin: option.value
                                });
                            }
                        });
                    }

                    // Product highlights
                    const highlights = [];
                    const highlightElems = document.querySelectorAll('#feature-bullets li span.a-list-item');
                    highlightElems.forEach(span => {
                        const text = span.innerText.trim();
                        if (text && !text.includes('See more product details')) {
                            highlights.push(text);
                        }
                    });

                    // Technical details
                    const technicalDetails = {};
                    const techSelectors = [
                        '#productDetails_techSpec_section_1 tr',
                        '#tech tbody tr',
                        '#technicalSpecifications_section_1 tr'
                    ];

                    techSelectors.forEach(selector => {
                        const techRows = document.querySelectorAll(selector);
                        techRows.forEach(row => {
                            const th = row.querySelector('th');
                            const td = row.querySelector('td');
                            if (th && td) {
                                technicalDetails[th.innerText.trim()] = td.innerText.trim();
                            }
                        });
                    });

                    // Main product image
                    const mainImageElem = document.querySelector('#landingImage, #imgTagWrapperId img, #imageBlock img');
                    const mainImage = mainImageElem ? mainImageElem.src : '';

                    // Amazon's Choice from acBadge_feature_div
                    const isAmazonChoice = document.querySelector('#acBadge_feature_div .mvt-ac-badge-rectangle') !== null;

                    // Service features from iconfarmv2
                    const serviceFeatures = [];
                    const iconElems = document.querySelectorAll('#icon-farm-container .icon-container');
                    iconElems.forEach(icon => {
                        const nameAttr = icon.getAttribute('data-name');
                        const contentElem = icon.querySelector('.icon-content a, .icon-content');
                        if (contentElem) {
                            serviceFeatures.push({
                                type: nameAttr,
                                text: contentElem.innerText.trim()
                            });
                        }
                    });

                    return {
                        title: title,
                        brand: brand,
                        rating: rating,
                        review_count: reviewCount,
                        price: price,
                        original_price: originalPrice,
                        discount_percentage: discount,
                        savings: originalPrice && price ? originalPrice - price : null,
                        exchange_offer: exchangeOffer,
                        stock_status: stockStatus,
                        in_stock: inStock,
                        delivery_time: deliveryTime,
                        deliver_to: deliverTo,
                        ships_from: shipsFrom,
                        sold_by: soldBy,
                        payment_info: paymentInfo,
                        gift_options: giftOptions,
                        colors: colors,
                        sizes: sizes,
                        highlights: highlights,
                        technical_details: technicalDetails,
                        main_image: mainImage,
                        is_amazon_choice: isAmazonChoice,
                        service_features: serviceFeatures,
                        url: window.location.href
                    };
                }
            """)

            print(f"✅ Product: {product_data['title'][:60]}...")
            if product_data['price']:
                price_str = f"   Price: ₹{product_data['price']:,}"
                if product_data['savings']:
                    price_str += f" (₹{product_data['savings']:,} off)"
                print(price_str)
            if product_data['rating'] and product_data['review_count']:
                print(f"   Rating: {product_data['rating']}⭐ ({product_data['review_count']:,} reviews)")
            print(f"   Stock: {product_data['stock_status']}")
            if product_data['delivery_time']:
                print(f"   Delivery: {product_data['delivery_time'][:50]}...")
            if product_data['colors']:
                print(f"   Colors: {len(product_data['colors'])} options")
            if product_data['sizes']:
                print(f"   Sizes: {len(product_data['sizes'])} options")
            if product_data['is_amazon_choice']:
                print("   🏆 Amazon's Choice")

            result = {
                "success": True,
                "product": product_data
            }

        except Exception as e:
            print(f"❌ Error viewing product: {e}")
            result = {
                "success": False,
                "error": str(e)
            }

        await context.close()
        await browser.close()
        await pw.stop()

        return result

    async def add_to_cart(self, product_urls):
        """
        BUG: Robot detection.
        Add one or multiple products to the shopping cart.

        Takes a product URL or list of URLs and adds one unit of each product to the cart.
        Processes multiple products efficiently by reusing the same browser session.

        Args:
            product_urls: Single product URL (string) or list of product URLs
                         Full URLs, ASINs, or tracking URLs supported

        Returns:
            Dict with success status and list of results for each product

        Example:
            # Single product
            await add_to_cart("https://www.amazon.in/dp/B0CHX1W1XY")

            # Multiple products
            await add_to_cart([
                "https://www.amazon.in/dp/B0CHX1W1XY",
                "B0CHX2F5QT",
                "/dp/B0CHX6NQMD"
            ])
        """
        # Handle single URL or list of URLs
        if isinstance(product_urls, str):
            product_urls = [product_urls]

        # Ensure all URLs are full URLs
        import re
        full_urls = []
        for url in product_urls:
            if not url.startswith('http'):
                # Check if it's just an ASIN
                if re.match(r'^[A-Z0-9]{10}$', url):
                    url = f"https://www.amazon.in/dp/{url}"
                else:
                    url = f"https://www.amazon.in{url}"
            else:
                # Extract ASIN from tracking URLs
                asin_match = re.search(r'/dp/([A-Z0-9]{10})', url)
                if asin_match:
                    url = f"https://www.amazon.in/dp/{asin_match.group(1)}"
            full_urls.append(url)

        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        results = []
        print(f"🛒 Adding {len(full_urls)} product(s) to cart...")

        for idx, product_url in enumerate(full_urls, 1):
            try:
                print(f"  [{idx}/{len(full_urls)}] Opening: {product_url}")
                await page.goto(product_url, wait_until="domcontentloaded")

                # Wait a bit for page to stabilize
                await page.wait_for_timeout(1000)

                # Check if already in cart (quantity stepper visible)
                already_in_cart = await page.locator('.sc-quantity-stepper').count() > 0

                if already_in_cart:
                    print(f"  ℹ️ Product {idx} already in cart, incrementing quantity...")

                    # Click the increment button
                    clicked = await page.evaluate("""
                        () => {
                            const incrementBtn = document.querySelector('[data-action="a-stepper-increment"]');
                            if (incrementBtn) {
                                incrementBtn.click();
                                return true;
                            }
                            return false;
                        }
                    """)

                    if clicked:
                        await page.wait_for_timeout(1500)
                        print(f"  ✅ Product {idx} quantity increased!")
                        results.append({
                            "success": True,
                            "url": product_url,
                            "index": idx - 1,
                            "action": "quantity_increased",
                            "already_in_cart": True
                        })
                    else:
                        print(f"  ⚠️ Could not find increment button for product {idx}")
                        results.append({
                            "success": False,
                            "url": product_url,
                            "index": idx - 1,
                            "error": "Increment button not found"
                        })
                else:
                    # Add to cart using the add button
                    print(f"  ➕ Adding product {idx} to cart...")

                    clicked = await page.evaluate("""
                        () => {
                            // Try multiple selectors for the Add to Cart button
                            const selectors = [
                                '#add-to-cart-button',
                                'input[name="submit.add-to-cart"]',
                                '#submit\\\\.add-to-cart input',
                                '.a-button-primary input[value="Add to Cart"]'
                            ];

                            for (const selector of selectors) {
                                const btn = document.querySelector(selector);
                                if (btn) {
                                    btn.click();
                                    return true;
                                }
                            }
                            return false;
                        }
                    """)

                    if clicked:
                        # Wait for the modal/popup to appear
                        await page.wait_for_timeout(2000)

                        # Close the modal by clicking the X button or clicking outside
                        modal_closed = await page.evaluate("""
                            () => {
                                // Try to find and click the close button on the modal
                                const closeSelectors = [
                                    '[data-action="a-popover-close"]',
                                    '.a-popover-close',
                                    'button[aria-label="Close"]',
                                    '[data-csa-c-type="element"][data-csa-c-action="dismiss"]'
                                ];

                                for (const selector of closeSelectors) {
                                    const closeBtn = document.querySelector(selector);
                                    if (closeBtn) {
                                        closeBtn.click();
                                        return true;
                                    }
                                }

                                // If no close button found, press Escape
                                return false;
                            }
                        """)

                        if not modal_closed:
                            # Try pressing Escape to close modal
                            await page.keyboard.press('Escape')

                        await page.wait_for_timeout(500)

                        # Now check if item is in cart by going to cart page
                        # This is more reliable than checking for confirmation messages
                        print(f"  ✅ Product {idx} added to cart!")
                        results.append({
                            "success": True,
                            "url": product_url,
                            "index": idx - 1,
                            "action": "added_to_cart",
                            "already_in_cart": False
                        })
                    else:
                        print(f"  ⚠️ Could not find Add to Cart button for product {idx}")
                        results.append({
                            "success": False,
                            "url": product_url,
                            "index": idx - 1,
                            "error": "Add to Cart button not found"
                        })

            except Exception as e:
                print(f"  ❌ Error adding product {idx}: {e}")
                results.append({
                    "success": False,
                    "url": product_url,
                    "index": idx - 1,
                    "error": str(e)
                })

        # Save session (cart state) once at the end
        await context.storage_state(path=self.state)
        await context.close()
        await browser.close()
        await pw.stop()

        successful = sum(1 for r in results if r["success"])
        print(f"🎉 Successfully added/updated {successful}/{len(full_urls)} products in cart!")

        return {
            "success": successful > 0,
            "total_products": len(full_urls),
            "successful": successful,
            "failed": len(full_urls) - successful,
            "results": results
        }
    
    async def view_cart(self):
        """
        View all items currently in the Amazon cart.

        Opens the cart page, extracts each product's name, URL, quantity, and price,
        and returns them as a structured list. If the cart is empty, returns an empty list.

        Returns:
            dict: {
                "success": bool,
                "total_items": int,
                "subtotal": str or None,
                "items": [
                    {
                        "asin": str,
                        "title": str,
                        "url": str,
                        "image": str,
                        "quantity": int,
                        "price": str
                    },
                    ...
                ]
            }
        """

        print("🧺 Viewing cart...")

        pw, browser = await get_browser()
        context = await browser.new_context(storage_state=self.state)
        page = await context.new_page()

        await page.goto("https://www.amazon.in/gp/cart/view.html", wait_until="domcontentloaded")
        await page.wait_for_timeout(1500)

        # Detect empty cart
        empty_cart = await page.locator(".sc-your-amazon-cart-is-empty").count() > 0
        if empty_cart:
            print("🫙 Cart is empty.")
            await context.close()
            await browser.close()
            await pw.stop()
            return {
                "success": True,
                "total_items": 0,
                "subtotal": None,
                "items": []
            }

        print("🔍 Extracting cart items...")

        # Extract all cart item elements
        item_elements = await page.locator('div.sc-list-item[data-asin]').element_handles()

        items = []
        for idx, item in enumerate(item_elements, start=1):
            try:
                asin = await item.get_attribute("data-asin")
                title = await item.eval_on_selector(".sc-product-title", "el => el.innerText.trim()") or ""
                url = await item.eval_on_selector(".sc-product-link", "el => el.href") or ""
                image = await item.eval_on_selector(".sc-product-image", "el => el.src") or ""
                quantity = await item.get_attribute("data-quantity") or "1"

                # Extract price safely
                try:
                    price_whole = await item.eval_on_selector(".a-price-whole", "el => el.innerText") or ""
                    price_fraction = await item.eval_on_selector(".a-price-fraction", "el => el.innerText") or ""
                    price = f"₹{price_whole.strip()}.{price_fraction.strip()}" if price_whole else None
                except Exception:
                    price = None

                items.append({
                    "asin": asin,
                    "title": title,
                    "url": url,
                    "image": image,
                    "quantity": int(quantity),
                    "price": price
                })

                print(f"  ✅ [{idx}] {title[:50]}... — {quantity} × {price}")
            except Exception as e:
                print(f"  ⚠️ Error parsing item {idx}: {e}")

        # Extract subtotal safely
        try:
            subtotal_text = await page.eval_on_selector("#sc-subtotal-amount-activecart", "el => el.innerText.trim()")
        except Exception:
            subtotal_text = None

        await context.close()
        await browser.close()
        await pw.stop()

        print(f"🧾 Found {len(items)} item(s) in cart.")
        if subtotal_text:
            print(f"💰 Subtotal: {subtotal_text}")

        return {
            "success": True,
            "total_items": len(items),
            "subtotal": subtotal_text,
            "items": items
        }
```
